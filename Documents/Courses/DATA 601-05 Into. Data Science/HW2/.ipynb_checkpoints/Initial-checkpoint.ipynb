```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://www.30dayweather.com/en/d/japan/tokyo-to/tokyo/'
page = requests.get(url)
print(page)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
day = []
date = []
partic = []
high = []
low = []


page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

#    day
for j in soup.findAll('div', class_='text-center text-color fct-day-of-week'):
    day.append(j.get_text())

#    date
for k in soup.findAll('div', class_='text-center fct-day-of-month'):
    date.append(k.get_text())

#   high temperature
for l in soup.findAll('span', class_='center-block label label-danger label-temp'):
    high.append(l.get_text())
              
#   low temperature
for m in soup.findAll('span', class_='center-block label label-info label-temp'):
    low.append(m.get_text())

time = []
for i in date:
     time.append(i.strip())

data = pd.DataFrame({'Day':day,
                     'Date':time,
                     'High Temperature':high,
                     'Low Temperature':low,})

data
```

```python
data.to_csv('JP_Forecast.csv')
```
