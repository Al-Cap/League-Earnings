```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

url = 'https://www.esportsearnings.com/leagues'
page = requests.get(url)
print(page)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
leag = []
name = []
prize = []
ply = []
tour = []
ev = []

page = requests.get(url)


#for i in soup.findAll('h3', class_='detail_list_table_header'):
 #   leag.append(i.get_text())

for j in soup.findAll('td', class_='format_cell detail_list_player'):
    name.append(j.get_text())
    
for k in soup.findAll('td', class_='format_cell detail_list_prize'):
    prize.append(k.get_text())
    
for l in soup.findAll('td', class_='format_cell detail_list_prize'):
    ply.append(l.get_text())

for m in soup.findAll('td', class_='format_cell detail_list_prize'):
    tour.append(m.get_text())    
    
for n in soup.findAll('td', class_='format_cell detail_list_prize'):
    ev.append(n.get_text())
    

    
money = prize[0::4]
ply = ply[1::4]
tour = tour[2::4]
ev = ev[3::4]



#ply = map(lambda s: s.strip('Players'), ply)
#money = map(lambda s: s.strip('$'), money)
#tour = map(lambda s: s.strip('Tournaments'), ply)
#ev = map(lambda s: s.strip('Events'), money)


#print(len(money))
#len(leag)






table = pd.DataFrame({'league':name, 'earnings':money, 'players':ply, 'tournaments':tour, 'events':ev})
table
```

```python
table['players'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
table['tournaments'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
table['events'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')

def clean_currency(x):
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)

table['earnings'] = table['earnings'].apply(clean_currency).astype('float')


table
```

```python
table['earnings'] = table['earnings'].astype(float)
table['players'] = pd.to_numeric(table['players'],errors='coerce')
table['tournaments'] = pd.to_numeric(table['tournaments'],errors='coerce')
table['events'] = pd.to_numeric(table['events'],errors='coerce')
table.info()
print(table)
```
